# 插值方法

### 函数定义以及实现

#### Lagrange插值

```python
# Lagrange 
def lagrange_interpolation(x, y, x0):  
    n = len(x)  
    L = 0  # Lagrange插值函数
    for k in range(n):  
        lk = 1  
        for i in range(n):  
            if i != k:  
                lk *= (x0 - x[i]) / (x[k] - x[i])  # Lagrange基函数，直接将x0代入
        L += lk * y[k]  
    return L  
```

#### Newton插值  

```python
# Newton 
def newton_interpolation(x, y, x0):  
    n = len(x)  
    # 构建差商表
    f = np.zeros((n, n))  
    f[:, 0] = y  # 初始化第一列为y值，构造n-1阶差商
    for j in range(1, n):
        for i in range(j, n):
            f[i][j] = (f[i][j-1] - f[i-1][j-1]) / (x[i] - x[i-j]) 
            
    # 使用差商表计算N(x) = Σ (f[x0...xi] * wi(x))，使用秦九韶算法
    N = f[n-1][n-1] # 最后一个元素是插值多项式的常数项  
    for j in range(n-2, -1, -1): # 从最高次幂开始到最低次幂  
        N = N * (x0 - x[j]) + f[j][j]  
    return N 
```

### 正常数据组、龙格现象数据组

#### 正常数据组

```python
x = np.array([0.4, 0.5, 0.6, 0.7, 0.8, 0.9, 1.0, 1.1])  
y = np.array([-0.9163, -0.6931, -0.5108, -0.3567, -0.2231, -0.105, 0, 0.095])  

# Lagrange、Newton插值计算 
x0 = np.linspace(x[0], x[-1], 20)  # 将x取值区间分段  
y_lagrange = [lagrange_interpolation(x, y, xi) for xi in x0]  # Lagrange
y_newton = [newton_interpolation(x, y, xi) for xi in x0]  # Newton  
#插值结果
# print('data\n', x0)
# print('y_lagrage\n',y_lagrange)
# print('y_newton\n',y_newton)

# 绘图
plt.figure(figsize=(8, 6)) 
plt.subplot(2,1,1)
# plt.plot(x0, y_lagrange, label='Lagrange Interpolation') # Lagrange
plt.scatter(x0, y_lagrange, color = 'blue', label = 'Lagrange Interpolation')
plt.scatter(x, y, color='red', label='Data Points')  # 显示原始数据点  
plt.legend()  # 显示图例  

plt.subplot(2,1,2)
# plt.plot(x0, y_newton, label='Newton Interpolation') # Newton 
plt.scatter(x0, y_newton, color = 'green', label = 'Newton Interpolation')
plt.scatter(x, y, color='red', label='Data Points')  # 显示原始数据点  
plt.legend()  # 显示图例  
plt.show()
```

### 曲线拟合的最小二乘法

```python
def curve_fit(x,y,n):
    sum_xy = sum(x*y)
    sum_x_square = sum(x*x)
    sumx = sum(x)
    sumy = sum(y)
    b = (n*sum_xy - sumx*sumy)/(n*sum_x_square - sumx**2)
    a = (sumy - b*sumx)/n
    print(f'y = a（{a}）+ b({b}) x')
    
n = 5
x = np.array([165, 123, 150, 123, 141])
y = np.array([187, 126, 172, 125, 148])
curve_fit(x,y,n)
```

# 数值积分与微分

## 复化梯形公式（变步长算法）

### 不使用递推公式

```python
# 被积函数f(x)
def f(x):
    return x**2 + 2*x
```

```python
# T = h*[f(a) + 2Σf(xi) + f(b)]/2 ,i从1到n-1
def compound(a, b, N):
    h = (b - a) / N  # 初始步长  
    x = np.linspace(a, b, N+1)  # 在[a, b]之间均匀地取N+1个点  
    y = [f(i) for i in x] 
    integral = h/2 * (y[0] + 2*np.sum(y[1:-2]) +y[-1])  
    print('初值: '+str(integral))
    # 变步长迭代  
    cnt = 1
    while True:  
        N *= 2  # 区间加倍
        h /= 2 # 步长减半
        x = np.linspace(a, b, N+1)  
        y = [f(i) for i in x] 
        new_integral = h/2 * (y[0] + 2*np.sum(y[1:-2]) +y[-1])  # 使用复化梯形公式计算积分
        if abs(new_integral - integral) < 0.0001:  # T2 - T1 精度达到要求
            break  
        integral = new_integral  # 更新积分值 
        print(f'迭代{cnt}次: '+str(integral)) # 输出每次迭代结果 
        cnt += 1
    return integral  
```

```python
# 计算 f(x) = x^2 + 2x 在[1,3]上的定积分
a = 1
b = 3
N = 2 # 区间分段数
print(compound(a,b,N))
```

### 使用递推公式

```python
# T2 = T1/2 + h[Σf(X(k+1/2))]/2 ,i从0到n-1
import numpy as np
def compound(a, b, N):
    h = (b - a) / N  # 初始步长  
    x = np.linspace(a, b, N+1)  # 在[a, b]之间均匀地取N+1个点  
    y = [f(i) for i in x] 
    integral = h/2 * (y[0] + 2*np.sum(y[1:-2]) +y[-1])   
    print('初值: '+str(integral))
    # 变步长迭代  
    cnt = 1
    while True:  
        x1 = x # 原来x
        x2 = [ x+ h/2 for x in x1[:-1]] #二分所得的x
        x_ = sorted(list(x1)+list(x2)) # 二分后全部x
        
        new_y = [f(x) for x in x2[:-1]] # 新增加的x对应的y值
        new_integral = 1/2 * integral + h/2 * np.sum(new_y)  
        if abs(new_integral - integral) < 0.0001:  # T2 - T1 精度达到要求
            break  
        integral = new_integral # 更新积分值 
        print(f'迭代{cnt}次: '+str(integral)) # 输出每次迭代结果 
        cnt +=1
        h/=2
        x = x_
    return integral  
```

```python
a = 1
b = 3
N = 2 # 区间分段数
print(compound(a,b,N))
```

## 改进欧拉公式

```python
def f(x, y): #f(x,y)
    return -y  

def improve_euler(x0, y0, h, n):
    # (x0,y0)为初始值坐标，h为步长，n为总步数
    x = np.zeros(n+1)
    y = np.zeros(n+1)
    x[0] = x0; y[0] = y0
    for i in range(0,n): 
        y[i+1] = y[i]+h*(f(x[i],y[i])+f(x[i+1], y[i]+h*f(x[i],y[i])))/2
    print(y)
```

```python
improve_euler(0,1,0.2,10)
```

## 四阶龙格库塔方法

```python
def f(x,y): #f(x,y)
    return -y
def runge(x0, y0, h, n):
    x = np.zeros(n+1)
    y = np.zeros(n+1)
    x[0] = x0; y[0] = y0
    for i in range(n):
        K1 = f(x[i],y[i])
        K2 = f(x[i]+h/2, y[i]+h*K1/2)
        K3 = f(x[i]+h/2, y[i]+h*K2/2)
        K4 = f(x[i]+h, y[i]+h*K3)
        y[i+1] = y[i] + h*(K1+2*K2+2*K3+K4)/6
    print(y)
```

```python
runge(0,1,0.4,5)
```

# 方程求根的数值方法

## 二分法

```python
def f(x): #f(x)
    return 2*x
def binary_split(a,b,min_distance):
    if f(a)*f(b) > 0:
        print('算法失效')
    while(True): 
        x = (a+b)/2
        if b-a < min_distance: #达到精度
            print(f'根区间({a},{b})')
            break
        elif f(x) == 0: #恰好得到精确解
            print('精确解：',x)
            break
        #二分迭代
        else:
            if f(x)*f(a) < 0:
                b = x
            else:
                a = x
```

```python
# 二分法求 y = 2x 的根
binary_split(-1,1.1,0.1)
```

## newton迭代法

```python
def f(x):#f(x)
    return x**3 -x -1
def df(x): #f'(x)
    return 3*x**2 -1

def newton_iter(x0,min): #x0为初值，min为精度
    x = x0
    iter_cnt = 0
    while abs(f(x))> min:
        x = x - f(x)/df(x)
        iter_cnt += 1
        print(f'第{iter_cnt}次迭代: ',x)
    print(f'迭代{iter_cnt}次，解得方程的根: ',x)
```

```python
newton_iter(12,0.5)
```

# 线型方程组的数值解法

```python
#测试方程组
a = np.array([[10,-1,-2], #a为系数矩阵
     [-1,10,-2],
     [-1,-1,5]])
b = np.array([7.2,       
     8.3,
     4.2])
```

## Jacobi法求n阶线性方程组

```python
def jacobi(a,b,min):
    n = 0 #迭代次数
    x = np.zeros(3) #初始化[x1,x2,x3]为0
    x_before = np.zeros(3)
    while n < 50:
        n+=1
        print('----------变化过程----------')
        print('迭代开始前: ',x)
        for i in range(3):
            sum = 0
            for j in range(3):
                if i != j:
                    sum += a[i][j] * x_before[j]
            x[i] = (b[i] - sum)/a[i][i]
            print(x)
        #判断精度
        if abs(max(x-x_before)) < min:
            print(f'迭代{n}次,精度达到要求:')
            print('上一次迭代: ',x)
            print('最终结果: ',x_before)
            return
        x_before = x.copy()
        
    print('迭代失败')
    
jacobi(a,b,0.001)
```

## Gauss-Seidel法求n阶方程组

```python
def gauss_seidel(a, b, min):
    n = 0
    x =np.zeros(3)
    x_before = np.zeros(3)
    while(n<50):
        n += 1
        print('------------变化过程-----------')
        print('迭代开始前：',x)
        for i in range(3):
            sum = 0
            for j in range(3):
                if i==j:
                    continue
                if j < i:
                    sum += a[i][j]*x[j]
                else:
                    sum += a[i][j]*x_before[j]
            x[i] = (b[i]-sum)/a[i][i]
            print(x)
        
        #判断精度
        if abs(max(x-x_before)) < min:
            print(f'迭代{n}次,精度达到要求:')
            print('上一次迭代: ',x)
            print('最终结果: ',x_before)
            return
        x_before = x.copy()
    print('迭代失败')

gauss_seidel(a,b,0.001)
```
